```python
race_anchor_templates = df[
    (df["Demographic Group"] == "Race") &
    (df["Demographic Attribute"].isin([
        "a White person",
        "White people"
    ]))
]
```

```python
race_anchor_templates
```

```python
race_anchor_templates.shape
race_anchor_templates["Demographic Attribute"].value_counts()
```

```python
for s in race_anchor_templates["Augmented Template"]:
    print(s)
```

```python
translation_df = race_anchor_templates[[
    "Augmented Template"
]].drop_duplicates().reset_index(drop=True)

translation_df = translation_df.rename(
    columns={"Augmented Template": "text_en"}
)

translation_df.head(5)
```

```python
output_path = "../../dataset/translation/race_anchor_templates_en.csv"
translation_df.to_csv(output_path, index=False, encoding="utf-8")

print(f"Saved {len(translation_df)} templates to {output_path}")
```
