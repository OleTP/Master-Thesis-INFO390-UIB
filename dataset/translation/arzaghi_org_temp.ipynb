```python
import pandas as pd
import numpy as np
```

### Arzaghi et al., (2024) - Augmented_Templates.csv

```python
df = pd.read_csv('../../../helper_datasets/mina_arzaghi/Augmented_Templates.csv')  # Read only first 10,000 rows
df.head(10)
```

### The Original Templates

```python
original_templates_df = df[df["Change"].isin(["always", "never"])][["Augmented Template", "Change"]]
original_templates_df
```

```python
num_sentences = len(original_templates_df)
print(num_sentences)
```

```python
num_unique = original_templates_df["Augmented Template"].nunique()
print(num_unique)
```

```python
original_templates_df = original_templates_df.rename(
    columns={"Augmented Template": "template_text_en"}
)

original_templates_df.head(5)
```

```python
output_path = "../../../helper_datasets/translation/original_templates.csv"
original_templates_df.to_csv(output_path, index=False, encoding="utf-8")

print(f"Saved {num_sentences} templates to {output_path}")
```
